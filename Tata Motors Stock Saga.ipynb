# Tata Motors Stock Analysis: A Journey from 1991 to 2024

# Introduction:
Tata Motors Limited, an Indian multinational automotive company and a part of the Tata Group, has traversed a remarkable path in the stock market from 1991 to 2024. This report delves into the historical performance of Tata Motors' stock, examining key trends, milestones, and factors influencing its trajectory over the decades.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mplfinance as mpf
import warnings
warnings.filterwarnings('ignore')
```

# Stock Price Evolution:
Over the past three decades, Tata Motors' stock price has witnessed significant fluctuations, reflecting both internal operational dynamics and broader market conditions. From humble beginnings in the early 1990s, the stock gradually gained traction, propelled by strategic acquisitions, innovative product launches, and expanding global footprints. However, it also weathered challenges, including economic downturns, regulatory changes, and industry disruptions.

```python
df = pd.read_csv("/kaggle/input/tata-motors-1991-2024/TATAMOTORS.NS.csv")
```

# Financial Performance:
The financial performance of Tata Motors mirrored its stock price journey, with periods of robust growth interspersed with periods of volatility. Revenue growth, profit margins, and market share fluctuations played crucial roles in shaping investor sentiment and influencing stock valuations. Notably, the company's acquisition of British luxury carmaker Jaguar Land Rover in 2008 marked a significant milestone, expanding its portfolio and global presence but also introducing new complexities and challenges.

```python
df.head()
```

# Market Dynamics:
Throughout its history, Tata Motors navigated the complexities of the automotive industry, characterized by intense competition, evolving consumer preferences, and technological advancements. The company's ability to innovate, adapt, and anticipate market trends influenced its competitive positioning and long-term sustainability. Moreover, strategic partnerships, joint ventures, and investments in research and development bolstered Tata Motors' capabilities and resilience in the face of industry disruptions.

```python
sns.set_palette("Blues")

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], color='steelblue')
plt.title('Tata Motors Closing Prices (1991-2024)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks([])
plt.show()
```

# Investor Sentiment:
Investor sentiment towards Tata Motors fluctuated over the years, reflecting a blend of fundamental analysis, market speculation, and macroeconomic factors. While periods of optimism were fueled by strong financial results, product launches, and growth prospects, downturns were often triggered by concerns over profitability, debt levels, and macroeconomic headwinds. Additionally, factors such as regulatory changes, geopolitical tensions, and environmental considerations increasingly influenced investor perceptions and investment decisions.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Volume'], bins=50, color='royalblue')
plt.title('Distribution of Daily Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

plt.figure(figsize=(12, 6))
mpf.plot(df, type='candle', style='charles', title='Tata Motors Candlestick Chart', ylabel='Price', ylabel_lower='Volume', figratio=(12, 6))
plt.show()
```

```python
df['Year'] = df.index.year

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Year', y='Volume', estimator=sum, ci=None, color='cornflowerblue')
plt.title('Tata Motors Volume by Year')
plt.xlabel('Year')
plt.ylabel('Total Volume')
plt.xticks(rotation=90)
plt.show()
```

```python
df['Month'] = df.index.month

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Month', y='Close', estimator=np.mean, ci=None, color='lightskyblue')
plt.title('Tata Motors Monthly Average Closing Prices')
plt.xlabel('Month')
plt.ylabel('Average Closing Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Close', color='lightsteelblue')
plt.title('Tata Motors Daily Closing Prices Boxplot')
plt.ylabel('Closing Price')
plt.show()
```

# Outlook:
As Tata Motors embarks on its journey into the future, the company faces both opportunities and challenges in an increasingly dynamic and competitive landscape. Key focus areas include sustainable mobility solutions, digital transformation, and diversification strategies to mitigate risks and capture emerging growth opportunities. Furthermore, enhancing operational efficiencies, strengthening customer engagement, and fostering innovation will be critical drivers of long-term value creation and shareholder returns.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Close', kde=True, color='dodgerblue')
plt.title('Distribution of Tata Motors Daily Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Year'] = df.index.year

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Close', estimator=np.mean, ci=None, color='royalblue')
plt.title('Tata Motors Yearly Trend of Closing Prices')
plt.xlabel('Year')
plt.ylabel('Average Closing Price')
plt.xticks(rotation=45)
plt.show()
```
